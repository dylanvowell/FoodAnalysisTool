```python
def get_matching_food_items(food):
    matches = process.extract(food, aggregated_data['food_item'], limit=None)
    
    #had to use regex to further filter the matches list since fuzzywuzzy wasn't doing the best job. 
    filtered_matches = [match for match in matches if re.search(r'\b' + re.escape(food) + r'\b', match[0], flags=re.IGNORECASE)]
    
    #this is basically saying that if the length of the filtered_matches is greater than 1, then list all of the potential matches and have the user select the index number of the item they want
    if len(filtered_matches) > 1:
        print("Looks like there are a few potential matches for your food item:")
        for i, match in enumerate(filtered_matches, start=1):
            print(f"{i}. {match[0]}")  # Print both index number and item name
        selection = int(input("Enter the number corresponding to your choice on the list above: "))
        selected_item_name = filtered_matches[selection - 1][0] if 0 < selection <= len(filtered_matches) else None
        print(f"You selected: {selected_item_name}")
        return selected_item_name
    elif filtered_matches:
        return filtered_matches[0][0]
    else:
        return None

food_query = input("Enter a food product: ")
matching_food_items = get_matching_food_items(food_query)
food_item_table
```

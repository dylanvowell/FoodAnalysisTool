```python
from fuzzywuzzy import process
```

```python
def get_nutrients_with_fuzzy_matching_and_units(food):
    # Perform fuzzy matching to find the closest match to the input food item
    match = process.extractOne(food, foundation_food_df['food_item'])
    closest_food = match[0]
    
    # Filter the DataFrame for the closest matched food item
    food_data = foundation_food_df[foundation_food_df['food_item'] == closest_food]
    
    nutrient_info = {}
    
    # Populate the dictionary with nutrients, amounts, and units
    for index, row in food_data.iterrows():
        nutrient = row['component']
        amount = row['amount']
        unit = row['unit']
        nutrient_info[nutrient] = {'amount': amount, 'unit': unit}
    
    return nutrient_info
```

```python
# Input from user
food_item = input("Enter a food product: ")
nutrient_info = get_nutrients_with_fuzzy_matching_and_units(food_item)

# Sort the dictionary by keys alphabetically
nutrient_info_sorted = dict(sorted(nutrient_info.items()))

# Create a DataFrame from the sorted dictionary
nutrient_df = pd.DataFrame(nutrient_info_sorted).T

# Print the DataFrame
print(f"Nutrients contained in {food_item}:")
nutrient_df
```

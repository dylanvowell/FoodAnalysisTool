```python
aggregated_data = foundation_food_df.groupby('food_item').agg({
    'component': list,
    'amount': list,
    'unit': list
}).reset_index()

aggregated_data
```

```python
# Assuming 'food_item_name' is the name of the food item you want to retrieve
food_item_name = 'Hummus, commercial'

# Filter the DataFrame based on the 'food_item' column
food_item_data = aggregated_data[aggregated_data['food_item'] == food_item_name]

# Create a new DataFrame for the specific food item
food_item_table = pd.DataFrame({
    'Nutrient Component': food_item_data['component'].iloc[0],
    'Amount': food_item_data['amount'].iloc[0],
    'Unit': food_item_data['unit'].iloc[0]
})

# Print the table-like output for the specific food item
print(f"Nutrients contained in {food_item_name}:")
print(food_item_table)
```
